# DATA ANALYSIS AND VISUALISATIONS

```python
cancelled_perc = df['is_canceled'].value_counts(normalize = True)
cancelled_perc

# now we are making the plot to visualize the percentage of cencelled bookings in both hotels with matplotlib library.

plt.figure(figsize =(5,4))
plt.title('Reservation Status Counts ')
plt.bar(['Not Canceled', 'Canceled'], df['is_canceled'].value_counts(), edgecolor= 'yellow', width = 0.7)
plt.show()
plt.savefig('Reservation Status Counts.jpg')

# In the provided code snippet, the normalize parameter in the value_counts() function is set to True....
# This parameter controls whether the returned counts are normalized or not.
# When normalize=True, the value_counts() function will return the relative frequencies or proportions......
# ....of each unique value in the 'is_canceled' column as a percentage of the total number of observations in the DataFrame.
```

 NOW WE WILL SEE THAT WHICH HOTEL HAS MORE CANCELLLATION RATIOS AND BY HOW MUCH SO WE HAVE DO SOMTHING ONLY IN THE 'IS_CANCELED' 
 COLUMN WITH THE HELP OF VISUALISATION IN SEABORN LIBRARY OF PYTHON 

```python
plt.figure(figsize = (8,4))

# since this is the categorical data which is 'is cancelled' and 'not cancelled' so there are plots for categorical datas in 
# seaborn library we will use here the count plot 

# fb vss bc
ax1= sns.countplot(x='hotel', hue= 'is_canceled', palette= 'Blues', data= df )
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title('Reservation Status in different hotels', size= 20)
# plt.ylabel("number of reservations")
# plt.xlabel('hotel')

plt.legend(['not canceled' , 'canceled'])
plt.show()
plt.savefig('Reservation Status in different hotels.jpg')
```

```python
# from above we can observe that resort hotels have less bookings as compared to city hotels but in resort hotels 
# cancellation rates are low because they give better facilities i think an in city hotels facilities and other things
# are not quuite good thats why cancellation rates are high,

# so conclution is that city hotels need to be look more 
```

```python
df.head()
# adr = average daily rate
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize= True)

# we are fetching that how much percentage of bookings are getting cancelled in resort hotels 
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize= True)

# we are fetching that how much percentage of bookings are getting cancelled in city hotels
```

```python
resort_hotel= resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel= city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize =(20,8))
plt.title('Average daily rate in City and Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label= 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label= 'City Hotel')
plt.legend( fontsize =20)
plt.show()
plt.savefig('ADR in city and resort hotels.jpg')


# it clears ours hypothesis that city hotel price is less than resort hotel in some days 
```

```python
df['month']  =  df['reservation_status_date'].dt.month
plt.figure(figsize = (20,8))
ax2= sns.countplot(x = 'month' ,hue = 'is_canceled' ,data= df, palette = 'bright')
plt.legend(['not_canceled', 'canceled'])
plt.title('Reservation Status Per Month', size = 30 )
ax2.set_xlabel('Months', size = 25)
ax2.set_ylabel('Total number of reservations ', size = 20)
plt.show()
plt.savefig('Reservation Status per month.jpg')
```

```python
# plotting average daily rate for each month 
```

```python
canceled_reservations_per_month = df[df['is_canceled'] == 1].groupby('month')['is_canceled'].count()
not_canceled_reservations_per_month = df[df['is_canceled'] == 0].groupby('month')['is_canceled'].count()
perc = canceled_reservations_per_month / (canceled_reservations_per_month + not_canceled_reservations_per_month)
perc*100 

# clearly august have minimum cancellation and january have maximum number of calcellations
```

```python

plt.figure(figsize=(20, 8))
plt.title('ADR per month for Canceled Reservations', fontsize=30)

# Filter the DataFrame to include only canceled reservations
canceled_df = df[df['is_canceled'] == 1]

# Group by month and calculate the sum of ADR for each month
adr_per_month = canceled_df.groupby('month')['adr'].sum().reset_index()

# Sort the DataFrame by month (optional, for better visualization)
adr_per_month = adr_per_month.sort_values(by='month')

# Create the bar plot
sns.barplot(x='month', y='adr', data=adr_per_month)

plt.xlabel('Months', fontsize=20)
plt.ylabel('Total ADR for Canceled Reservations', fontsize=20)

plt.show()
plt.savefig('ADR per month for cancelled reservation.jpg')




       
```
